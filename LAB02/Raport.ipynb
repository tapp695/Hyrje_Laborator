## Çfarë është Jupyter Notebook?

Jupyter Notebook është një mjedis interaktiv që lejon kombinimin e:
- Tekstit
- Kodit Python
- Formulave matematikore
- Grafikëve dhe rezultateve

Notebook përbëhet nga qeliza (cells):
- Markdown (tekst)
- Code (Python)


## Struktura e një Notebook

Një notebook zakonisht ndjek këtë strukturë:

- Titulli
- Objektivi
- Teoria
- Metoda
- Analiza
- Rezultatet
- Përfundimi

## Analizë e thjeshtë me Python


```python
import numpy as np
import matplotlib.pyplot as plt

time = np.array([0,1,2,3,4])
position = np.array([0,5,20,45,80])

plt.plot(time, position, 'o-')
plt.xlabel("Time (s)")
plt.ylabel("Position (m)")
plt.title("Motion Experiment")
plt.grid()
plt.show()
```
